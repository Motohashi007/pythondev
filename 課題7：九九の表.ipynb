```python
#@title 九九のリストを作成する
list_add = []
for v in range(1, 10):
    list_inner = []
    for h in range(1, 10):
        list_inner.append(v * h)
    list_add.append(list_inner)
print(list_add)
```

```python
#@title 九九の表を表示する
for a in range(0, 9):
    for b in range(0, 9):
        print(f"{list_add[a][b]:3d}", end="")
    print("")
```
