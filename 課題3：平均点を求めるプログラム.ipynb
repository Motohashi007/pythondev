```python
count = 0
score_sum = 0

while True:
    score = int(input("点数を入力してください："))

    if score == -1:
        print(f"点数を入力してください：{score}")
        break  
    
    else:
        print(f"点数を入力してください：{score}")
        count += 1
        score_sum = score_sum + score 
        
        
    average = round(score_sum / count ,1)

print(f"{count}人のテストの平均点は{average}点です")
```
