```python
num1 = input("割られる数を入力してください：")
num2 = input("割る数を入力してください：")

try:
    a = int(num1)
    b = int(num2)

    print(f"割られる数を入力してください：{num1}")
    print(f"割る数を入力してください：{num2}")

    c = a / b

    print(f"{a} ÷ {b} = {c}")

except ValueError:
    print(f"割られる数を入力してください：{num1}")
    print("エラー：数値を入力してください")

except:
    print("エラー：0で割り算しないでください") 

print("処理を終了します")
```
