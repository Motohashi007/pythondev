```python
excluding_tax = int(input("税抜価格を入力してください："))
```

```python
price = round(excluding_tax * 1.1)
price
```

```python

if price >= 2000:
    print(f"税抜価格を入力してください：{excluding_tax}")
    print("送料は無料です")
    print(f"送料込みの価格は{price}円です")
    
else:
    print(f"税抜価格を入力してください：{excluding_tax}")
    print("送料として350円かかります")
    print(f"送料込みの価格は{price+350}円です")
```
